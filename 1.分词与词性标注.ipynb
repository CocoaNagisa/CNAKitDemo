```python
import pandas as pd
import re
import jieba
```

```python
#  清理空格以及爬虫带来的编码问题
def clean(content):
    content = content.replace(' ','').replace('\xa0','').replace('\u3000','').replace('\ue010','')
    return content
```

```python
data_name = '沈阳市科学技术局'  # '沈阳市科学技术局'  '沈阳市人民政府' '沈阳市工业和信息化局'
```

```python
data = pd.read_table('{}.txt'.format(data_name), sep='\t', header=None)
data.columns = ['title','content']
data['content'] = data['content'].apply(clean)  #  清理爬虫带来的编码问题
data
```

```python
from LAC import LAC
lac = LAC(mode='seg')
def Seg(content):
    if content == 'None':
        return 'None'
    seg_result = lac.run(content)
    return '/ '.join(seg_result)
data['Seg'] = data['content'].apply(Seg)
data
```

```python
lac = LAC(mode='lac')
def Lac(content):
    if content == 'None':
        return 'None'
    lac_result = lac.run(content)
    lac_result = [x + '(' + y + ')' for x, y in zip(lac_result[0],lac_result[1])]
    return '/ '.join(lac_result)
data['Lac'] = data['content'].apply(Lac)
data
```

```python
data.to_csv(r'data/{}/{}.csv'.format(data_name,data_name),index=False, encoding='utf-8')
```
