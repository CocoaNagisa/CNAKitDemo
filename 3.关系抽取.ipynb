```python
import re
import pandas as pd
```

```python
data_name = '沈阳市工业和信息化局'  # '沈阳市科学技术局'  '沈阳市人民政府' '沈阳市工业和信息化局'
```

```python
data = pd.read_csv(r'data/{}/NER_{}.csv'.format(data_name,data_name), encoding='utf-8', header=0)
data
```

## 标题：(4关系+4属性) 三元组抽取

```python
head_entity = data_name
def title_relation_extraction(publish_time, TongZhi, JieDu, DaFu, PiFu):

    if TongZhi != 'None':
        triple_0 = (head_entity, '发布', TongZhi)   
        triple_1 = (TongZhi, '发布时间', publish_time)
        return str( (triple_0,triple_1) )

    elif JieDu != 'None':
        triple_0 = (head_entity, '解读', JieDu)   
        triple_1 = (JieDu, '解读时间', publish_time)
        return str( (triple_0,triple_1) )
    
    elif DaFu != 'None':
        triple_0 = (head_entity, '答复', DaFu)   
        triple_1 = (DaFu, '答复时间', publish_time)
        return str( (triple_0,triple_1) )
    
    elif PiFu != 'None':
        triple_0 = (head_entity, '批复', PiFu)
        triple_1 = (PiFu, '批复时间', publish_time)
        return str( (triple_0,triple_1) )

    else:
        return 'None'

data["title_relation"] = data.apply(lambda row:title_relation_extraction(row["publish_time"],row["TongZhi"],row["JieDu"],
                                                                         row["DaFu"],row["PiFu"]),axis=1)
data[["title_relation"]]
```

```python
title_relation = data[data["title_relation"] != 'None']["title_relation"].tolist()
title_relation
```

```python
len(title_relation)
```

```python
with open(r'data/{}/title_triple_{}.txt'.format(data_name,data_name), 'w', encoding='utf-8') as f:
    for triple in title_relation:
        triple = eval(triple)      
        f.write(str(triple[0]) + '\n')
        f.write(str(triple[1]) + '\n')
```

## 内容：联系人、联系方式 三元组抽取

```python
def contact_relation_extraction(contact_person, contact_number, TongZhi, JieDu, DaFu, PiFu):
    if TongZhi != 'None':
        triple = []
        if contact_person != 'None':
            triple_0 = (TongZhi, '联系人', contact_person)
            triple.append(triple_0)
        if contact_number != 'None':
            triple_1 = (TongZhi, '联系电话', contact_number)
            triple.append(triple_1)
        
        if len(triple) > 0: 
            return str(triple)
        else:
            return 'None'
        
    elif JieDu != 'None':
        triple = []
        if contact_person != 'None':
            triple_0 = (JieDu, '联系人', contact_person)
            triple.append(triple_0)
        if contact_number != 'None':
            triple_1 = (JieDu, '联系电话', contact_number)
            triple.append(triple_1)
        
        if len(triple) > 0: 
            return str(triple)
        else:
            return 'None'
        
    elif DaFu != 'None':
        triple = []
        if contact_person != 'None':
            triple_0 = (DaFu, '联系人', contact_person)
            triple.append(triple_0)
        if contact_number != 'None':
            triple_1 = (DaFu, '联系电话', contact_number)
            triple.append(triple_1)
        
        if len(triple) > 0: 
            return str(triple)
        else:
            return 'None'
    
    elif PiFu != 'None':
        triple = []
        if contact_person != 'None':
            triple_0 = (PiFu, '联系人', contact_person)
            triple.append(triple_0)
        if contact_number != 'None':
            triple_1 = (PiFu, '联系电话', contact_number)
            triple.append(triple_1)
        
        if len(triple) > 0: 
            return str(triple)
        else:
            return 'None'
    
    else:
        return 'None'

data["contact_relation"] = data.apply(lambda row:contact_relation_extraction(row["contact_person"],row["contact_number"],
                                                                             row["TongZhi"],row["JieDu"],row["DaFu"],row["PiFu"]),axis=1)
data[["contact_relation"]]
```

```python
contact_relation = data[data["contact_relation"] != 'None']["contact_relation"].tolist()
contact_relation
```

```python
len(contact_relation)
```

```python
with open('data/{}/contact_triple_{}.txt'.format(data_name,data_name), 'w', encoding='utf-8') as f:
    for triple in contact_relation:
        triple = eval(triple)
        for x in triple:
            f.write(str(x) + '\n')
```
